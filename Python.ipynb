```python
# Exception Handeling
- To ignore the execution error and continue to execute the code.
- try >> except >> else >> finally
- Error : NameError | TypeError | KeyError | ValueError | IndexError | AttributeError | AssertionError | ArgumentError | FileExistsError | FileNotFoundError | IndentationError
- SyntaxError is not handle by exception handeling

Syntax : 
try:
    statement 1 |_  Util this block do not get any error it will executes
    statement 2 |_
    statement 3 |
except:
        } _ Once try block have any error in pertcular statement since then except block will execute.

- If we want to see the error statement
import traceback
print(traceback.print_exc()) # in code

eg.:
import datetime
date = '01-11-1996'

formate = ['%d-%m-%Y', '%d/%m/%Y', '%Y/%m/%d', '%Y-%m-%d', '%Y %m %d', '%d %m %Y']


for frmt in formate:
    try:
        new = datetime.datetime.strptime(date, frmt).date()
        print(new)
        break
    except:
        continue



Syntax:
       try:
           |
           |_ No Error if Found 
           |_  Move to except.
        except:
            |
            |__ No error if got any move to execute else
            |
       else:
            |
            |__ if Found any then end the execution.
            |
       finally:
               This block always executes

* Raise
- Raise keyword is used to raise an error.
symntax:
       try:
          raise NameError('not Found')
        except:
              print('Wrong')
```
